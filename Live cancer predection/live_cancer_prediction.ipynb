```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# 🔹 Encode categorical variables (if not done already)
df_encoded = df.copy()
for col in df_encoded.select_dtypes(include=['object']).columns:
    if col != "Prediction":  # Exclude target variable
        df_encoded[col] = LabelEncoder().fit_transform(df_encoded[col])

# 🔹 Define features & target
Xtry = df_encoded.drop(columns=["Prediction"])  # Features
ytry = df_encoded["Prediction"]  # Target

# 🔹 Train a Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(Xtry, ytry)

# 🔹 Get feature importances
feature_importance = pd.DataFrame({
    "Feature": Xtry.columns,
    "Importance": model.feature_importances_
}).sort_values(by="Importance", ascending=False)

# 🔹 Select top 15 features
top_20_features = feature_importance.head(20)

# 🔹 Plot feature importance
plt.figure(figsize=(12, 6))
sns.barplot(x="Importance", y="Feature", data=top_20_features, palette="viridis")
plt.title("Top 20 Factors Influencing Disease Risk")
plt.xlabel("Feature Importance")
plt.ylabel("Feature")
plt.show()


# 🔹 Display the top 20 features
print(top_20_features)
```

```python
# Convert Prediction to numeric if it's not already
df["Prediction"] = df["Prediction"].map({"Yes": 1, "No": 0})  # Convert to 1 & 0 if needed

plt.figure(figsize=(12, 5))

# 🔹 Obesity vs Prediction
plt.subplot(1, 2, 1)
sns.countplot(x="Obesity", hue="Prediction", data=df, palette="coolwarm")
plt.xlabel("Obesity Level")
plt.ylabel("Count")
plt.title("Obesity vs. Liver Cancer Prediction")
plt.xticks(rotation=45)
plt.legend(title="Prediction", labels=["No Cancer", "Cancer"])

# 🔹 Diabetes vs Prediction
plt.subplot(1, 2, 2)
sns.countplot(x="Diabetes", hue="Prediction", data=df, palette="coolwarm")
plt.xlabel("Diabetes Status")
plt.ylabel("Count")
plt.title("Diabetes vs. Liver Cancer Prediction")
plt.xticks(rotation=45)
plt.legend(title="Prediction", labels=["No Cancer", "Cancer"])

plt.tight_layout()
plt.show()
```
